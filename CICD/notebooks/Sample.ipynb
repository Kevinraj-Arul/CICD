```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "John", "Cardiology", 45, 75000),
    (2, "Priya", "Neurology", 38, 82000),
    (3, "Ahmed", "Orthopedics", 50, 90000),
    (4, "Meena", "Pediatrics", 29, 60000),
    (5, "David", "Dermatology", 41, 72000),
    (6, "Suresh", "Radiology", 47, 88000),
    (7, "Anita", "Gynecology", 35, 70000),
    (8, "Rahul", "Oncology", 52, 95000),
    (9, "Sophia", "ENT", 33, 65000),
    (10, "Karthik", "General Medicine", 40, 68000)
]
columns = ["emp_id", "name", "department", "age", "salary"]
df = spark.createDataFrame(data, columns)
df.show()

# df.write.mode("overwrite").format("delta").saveAsTable('develop_data.employee_data.department')
# from pyspark.sql.functions import col
# df_filtered = df.filter((col("age") > 40) & (col("age") < 50))
# df_filtered.show()
```
